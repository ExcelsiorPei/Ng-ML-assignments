## 1.2 Implementation

### 1.2.1 Sigmoid Function

```python
def sigmoid(z):
    return 1. / (1 + np.exp(-z))
```

```python
x = np.linspace(-10, 10, 1000)
plt.figure(figsize=(10, 6))
plt.plot(x, sigmoid(x))
plt.show()
```

### 1.2.2 Cost function and gradient

```python
def costFunction(theta, X, y):
    m = len(y)
    theta = theta.reshape(theta.shape[0], 1) #后面有说明，theta.shape = (n,)
    h = sigmoid(np.dot(X, theta))
    cost = -1. / m * np.sum(np.multiply(y, np.log(h)) + np.multiply((1 - y), np.log(1 - h)))
    return cost
```

```python
data.insert(0, 'one', 1)
```

```python
colums = data.shape[1]
X = data.iloc[:, :colums-1]
y = data.iloc[:, colums-1:colums]
```

```python
X = np.array(X)
y = np.array(y)
theta = np.zeros(3)
```

**注：为使用opt的优化函数theta的shape必须是（n,）；gradient的返回值shape也必须是（n,）**

```python
X.shape, y.shape,theta.shape  #注：为使用opt的优化函数theta的shape必须是（n,）；gradient的返回值shape也必须是（n,）
```

```python
costFunction(theta, X, y)
```

** 梯度下降的步长函数 **

```python
def updateparameter(theta, X, y):
    m = len(y)
    theta = theta.reshape(theta.shape[0], 1)
    h = sigmoid(np.dot(X, theta))
    update = 1. / m * np.dot(X.T, (h - y))
    return update.squeeze()
```

```python
updateparameter(theta, X, y)
```

### 1.2.3 使用优化函数
根据文档中所写，并没用用传统的梯度下降而是采用的fminunc函数去自动进行优化，由于使用的Python所以换用其他方式

```python
import scipy.optimize as opt
result = opt.fmin_tnc(func=costFunction, x0=theta, fprime=updateparameter, args=(X,y))
result
```

```python
costFunction(result[0], X, y) #最终cost的值
```

### 1.2.4 Evaluating logistic regression

```python
def prediction(theta, X):
    theta = theta.reshape((theta.shape[0], 1))
    probability = sigmoid(np.dot(X, theta))
    return np.array([1 if i >= 0.5 else 0 for i in probability])  #记住用法if用在[]里面的简写方式
```

```python
predict = prediction(result[0], X).reshape(len(y), 1)
accuracy = len(predict[predict == y]) / len(y)
print("Accuracy = " + str(accuracy * 100) + "%")
```

# 2. Regularized logistic regression

## 1.1 Visualizing The Data

```python
path = 'ex2data2.txt' 
data = pd.read_csv(path, header=None, names=['Test1', 'Test2', 'Acception']) #header属性决定的是行名称，int或者list of ints
data.head()
```

```python
positive = data[data['Acception'].isin([1])]
negative = data[data['Acception'].isin([0])]

plt.figure(figsize=(10, 6))
plt.xlabel('Test1')
plt.ylabel('Test2')
plt.scatter(positive['Test1'], positive['Test2'], s=50, c='k', marker='+', label='Accepted')
plt.scatter(negative['Test1'], negative['Test2'], s=50, c='y', marker='o', label='Refused')
plt.legend(loc=1)
plt.show()
```

## 1.2Feature Mapping

```python
def featureMapping(x1, x2, power):
    #x1, x2为np数组
    #retrun DataFrame
    data = {}
    for i in range(power+1):
        for j in range(i+1):
            data['F{}{}'.format(i-j, j)] = np.power(x1, i-j) * np.power(x2, j)
    
    if type(x1) != np.ndarray:
        return pd.DataFrame(data, index=[0])
    return pd.DataFrame(data)
```

```python
x1 = np.array(data.Test1)
x2 = np.array(data.Test2)
data2_X = featureMapping(x1, x2, 6)
data2_X.head()
```

```python
def regularized_cost(theta, X, y, labmda=1):
    m = len(y)
    regularized_term = labmda / (2 * m) * np.sum(np.power(theta[1:], 2))
    
    return costFunction(theta, X, y) + regularized_term
```

**检查下列结果，与官方文档相同**

```python
theta = np.zeros(28)
cols= data.shape[1]
data2_y = np.array(data.iloc[:, cols-1:cols])
data2_X = np.array(data2_X)
cost = regularized_cost(theta, data2_X, data2_y)
cost
```

```python
def regularized_gradient(theta, X, y, labmda=1):
    m = len(y)
    regularized_term = theta[1:] * labmda / m  #theta0没有罚项
    regularized_term = np.concatenate([np.array([0]), regularized_term])  #刚开始写程序时都忘记了theta0要区别对待
    
    return updateparameter(theta, X, y) + regularized_term
```

```python
regularized_gradient(theta, data2_X, data2_y)
```

```python
 #利用和上一个例子相同的优化函数获得最终结果
result2 = opt.fmin_tnc(func=regularized_cost, x0=theta, fprime=regularized_gradient, args=(data2_X,data2_y))
result2
```

## 1.3结果预测

```python
theta = result2[0]
predict = prediction(theta, data2_X)
predict = predict.reshape(predict.shape[0], 1)
print("Accuracy = " + str(len(predict[predict == data2_y]) / len(data2_y) * 100) + '%')
```

## 1.4 画出决策边界 (不同$\lambda$情况下)

```python
def draw_boundary(theta):
    u = np.linspace(-1, 1.5, 50)
    v = np.linspace(-1, 1.5, 50)
    z = np.zeros((u.shape[0], v.shape[0]))
    power = 6
    for i in range(u.shape[0]):
        for j in range(v.shape[0]):
            z[i, j] = np.dot(featureMapping(u[i], v[j], power), theta)  #点乘后类型从DataFrame to np.ndarray
            
    #可以自行推到看np.meshgrid生成的函数坐标和上面这个循环所使用的坐标是否是转置的关系
            
    U, V = np.meshgrid(u, v) #生成u,v可能组合的所有平面的二维坐标（U,V）
    plt.figure(figsize=(12, 8))
    plt.scatter(positive['Test1'], positive['Test2'], s=50, c='k', marker='+')
    plt.scatter(negative['Test1'], negative['Test2'], s=50, c='y', marker='o')
    C = plt.contour(U, V, z.T, [0], colors='blue') #画三维等高线函数, 最后一个参数[0]代表只显示高度 = 0的那个等高线
    plt.xlabel('Test1')
    plt.ylabel('Test2')
    plt.legend(['Accepted', 'Rejected', 'Decision boundary'], loc=1)
    # plt.clabel(C, inline = True, fontsize = 10) #显示等高线数据
    plt.show()
```

## $\lambda$ = 1的情况， 结果还比较好

```python
draw_boundary(theta)
```

```python
def getheta(X, y, labmda):
    theta = np.zeros(28)
    result = opt.fmin_tnc(func=regularized_cost, x0=theta, fprime=regularized_gradient, args=(X,y,labmda))
    return result[0]
```
