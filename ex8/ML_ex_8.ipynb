## 1.1 高斯分布

```python
def estimate_gaussian(X):
    mu = X.mean(axis=0)
    sigma = X.var(axis=0)
    
    return mu, sigma
```

```python
mu, sigma = estimate_gaussian(X)
mu, sigma
```

我们还需要一种计算数据点属于正态分布的概率的方法。 幸运的是SciPy有这个内置的方法。

```python
from scipy import stats
x1_normal = stats.norm(mu[0], sigma[0])                    #生成两个高斯模型
x2_normal = stats.norm(mu[1], sigma[1])
```

## $P = p_1 * p_2$ 单变量高斯分布的处理方式

```python
x, y = np.mgrid[0:25:0.1, 0:25:0.1]
p1 = x1_normal.pdf(x)                              #值为x时属于此正态分布的概率就是p1
p2 = x2_normal.pdf(y)                              #值为y时属于此正态分布的概率就是p2


plt.figure(figsize=(12, 8))
plt.scatter(X[:, 0], X[:, 1])
C = plt.contour(x, y, p1 * p2, [0.000001, 0.00001, 0.0001, 0.001, 0.01], colors='black')   #画出,P分别为[0.00001, 0.0001, 0.001, 0.01, 0.5]的等高线
plt.clabel(C,inline=True,fontsize=10)                                            #每个等值线上标出P的值，貌似只能精确到3位小数
plt.show()
```

## 1.2 Selecting  the threshold
如果为 anomaly 则y = 1，正样本为异常样本

```python
# 计算f1_score，查看文档中的公式，我采用的遍历的方式，不过应该存在向量化的方式
def f1_score(y, y_hat):
    m = len(y)
    tp = fp = fn = 0
    
    for i in range(m):
        if y[i] == 1 and y_hat[i] == 1:
            tp += 1
        elif y[i] == 0 and y_hat[i] == 1:
            fp += 1
        elif y[i] == 1 and y_hat[i] == 0:
            fn += 1
    
    pre = tp / (tp + fp)
    rec = tp / (tp + fn)
    
#     print(pre, rec)
    return 2 * pre * rec / (pre + rec)
```

```python
#验证一下F1,手工计算，结果正确
f1_score([1, 0, 1], [1, 1, 0])
```

```python
def select_epsilon(X, Xval, yval):
    best_f1 = 0
    best_epsilon = 0
    predict = np.zeros(yval.shape)          #用于返回最终的预测结果
    
    mu, sigma = estimate_gaussian(X)
    x1_normal = stats.norm(mu[0], sigma[0])
    x2_normal = stats.norm(mu[1], sigma[1])
    
    x1_prob = x1_normal.pdf(Xval[:, 0])
    x2_prob = x2_normal.pdf(Xval[:, 1])
    P = x1_prob * x2_prob
    epsilon = np.linspace(P.min(), P.max(), num=10000)
    
    print(P.shape)
    
    for param in epsilon[1:]:                  #从第二个值开始是因为会存在除0的异常在计算F1时候，当然使用python的异常去解决也是可以的
        y_hat = (P < param).astype(int)
        f1 = f1_score(yval, y_hat)
        
        if  f1 > best_f1:
            best_f1 = f1
            best_epsilon = param
            predict = y_hat
    
    return best_epsilon, best_f1, predict
    
    
```

```python
epsilon, f1, _ = select_epsilon(X, Xval, yval)
```

```python
epsilon, f1
```

## 对训练集使用，查看什么样的点为异常点

```python
ptrain_x1 = x1_normal.pdf(X[:, 0])
ptrain_x2 = x2_normal.pdf(X[:, 1])
predict = np.where( (ptrain_x1 * ptrain_x2) < epsilon )                     #返回满足条件的下标组成的数组


plt.figure(figsize=(12, 8))
plt.scatter(X[:, 0], X[:, 1])
C = plt.contour(x, y, p1 * p2, [0.000001, 0.00001, 0.0001, 0.001, 0.01], colors='black')   #画出,P分别为[0.00001, 0.0001, 0.001, 0.01, 0.5]的等高线
plt.scatter(X[predict, 0], X[predict, 1], s=60, c='r')
plt.clabel(C,inline=True,fontsize=10)                                            #每个等值线上标出P的值，貌似只能精确到3位小数
plt.show()
```

### 虽然文档中说n = 100但是可以看出来实际上n = 10，文档说明有误

```python
def cost(params, Y, R, num_features):
    num_movies = Y.shape[0]
    num_users = Y.shape[1]
    X = params[:num_movies * num_features].reshape(num_movies, num_features)           #每个电影的特征
    theta = params[num_movies * num_features:].reshape(num_users, num_features)        #每个用户的特征系数
    
    J = 0
    error = (1. / 2) * np.power(np.dot(X, theta.T) - Y, 2) 
    J = np.sum(np.multiply(error, R))
    
    return J
```

```python
#测试正确性

users = 4
movies = 5
features = 3

y_1 = Y[:movies, :users]
r_1 = R[:movies, :users]
x_1 = X[:movies, :features]
theta_1 = Theta[:users, :features]
params_1 = np.concatenate((x_1.flatten(), theta_1.flatten()))      #flatten()与ravel()均可

cost(params_1, y_1, r_1, features)
```

```python
def gradient(params, Y, R, num_features):
    num_movies = Y.shape[0]
    num_users = Y.shape[1]
    X = params[:num_movies * num_features].reshape(num_movies, num_features)           #每个电影的特征
    theta = params[num_movies * num_features:].reshape(num_users, num_features)        #每个用户的特征系数
    
    error = np.multiply((np.dot(X, theta.T) - Y), R)
    dX = np.dot(error, theta)
    dtheta = np.dot(error.T, X)
    
    update = np.concatenate((dX.flatten(), dtheta.flatten()))
    return update
```

## 加入正则化

```python
def regularized_cost(params, Y, R, num_features, labmda):
    num_movies = Y.shape[0]
    num_users = Y.shape[1]
    X = params[:num_movies * num_features].reshape(num_movies, num_features)           #每个电影的特征
    theta = params[num_movies * num_features:].reshape(num_users, num_features)        #每个用户的特征系数
    
    J = 0
    error = (1. / 2) * np.power(np.dot(X, theta.T) - Y, 2) 
    J = np.sum(np.multiply(error, R))
    
    #正则项
    J += labmda / 2 * np.sum(np.power(theta, 2))
    J += labmda / 2 * np.sum(np.power(X, 2))
    
    return J
```

```python
def regularized_gradient(params, Y, R, num_features, labmda):
    num_movies = Y.shape[0]
    num_users = Y.shape[1]
    X = params[:num_movies * num_features].reshape(num_movies, num_features)           #每个电影的特征
    theta = params[num_movies * num_features:].reshape(num_users, num_features)        #每个用户的特征系数
    
    error = np.multiply((np.dot(X, theta.T) - Y), R)
    dX = np.dot(error, theta)
    dtheta = np.dot(error.T, X)
    
    #正则项
    dX += labmda * X
    dtheta += labmda * theta
    
    update = np.concatenate((dX.flatten(), dtheta.flatten()))
    
    return update
```

### 归一化评级，为了使得我们没有看过的电影有相对正确的评分预测，相当于对于没有评分的电影系统默认平均分

```python
Ymean = np.zeros((Y.shape[0], 1))
Ynorm = np.zeros(Y.shape)
#向量化不好处理，求均值要去掉非0的点
for i in range(Y.shape[0]):
#     Ymean[i] = np.sum(np.multiply(Y[i, :], R[i, :])) / len(R[i, :] != 0) 
    idx = np.where(R[i, :] != 0)[0]
    Ymean[i] = Y[i, idx].mean()
    Ynorm[i, idx] = Y[i, idx] - Ymean[i]                #只是那些打了分数的位置才减去均值，之前自己计算错误了，把所有位置都减去了均值
```
