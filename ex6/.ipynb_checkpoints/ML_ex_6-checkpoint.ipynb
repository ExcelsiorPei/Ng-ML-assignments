# 1 Support Vector Machines

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

## 1.1 Example Dataset 1

```python
mat = loadmat("data/ex6data1.mat")
data = pd.DataFrame(mat['X'], columns=['X1', 'X2'])
data['Y'] = mat['y']
print(data.shape)
data.head()
```

```python
positive = data[data['Y'] == 1] # data[data['y'].isin([1])] 这样表示结果一样，但是这个更常用
negative = data[data['Y'] == 0]
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(positive['X1'], positive['X2'], c='k', marker='+', label='positive')
plt.scatter(negative['X1'], negative['X2'], c='y', marker='o', label='negative')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```

# C = 1，直接从sklearn包中导入SVM
http://scikit-learn.org/stable/modules/generated/sklearn.svm.LinearSVC.html#sklearn.svm.LinearSVC

```python
from sklearn import svm
```

```python
svc1 = svm.LinearSVC(C=1, loss='hinge', max_iter=1000) #  ‘hinge’ is the standard SVM loss 
svc1
```

```python
svc1.fit(data[['X1', 'X2']], data['Y'])
svc1.score(data[['X1', 'X2']], data['Y'])
```

# C = 100

```python
svc2 = svm.LinearSVC(C=100, loss='hinge', max_iter=1000) #  ‘hinge’ is the standard SVM loss 
svc2.fit(data[['X1', 'X2']], data['Y'])
svc2.score(data[['X1', 'X2']], data['Y'])
```

```python
data['SVM1 Confidence'] = svc1.decision_function(data[['X1', 'X2']]) #决策边界的置信度
print(data['SVM1 Confidence'].head())
data['SVM2 Confidence'] = svc2.decision_function(data[['X1', 'X2']]) #决策边界的置信度，置信度越大说明离边界越远
print(data['SVM2 Confidence'].head())
```

## 这里可以明显看到左上角那个点在C=1 or 100条件下的不同结果
置信度越大说明离决策边界越远，颜色越深，此cmap代表正负置信度是双色

```python
plt.figure(figsize=(18, 6))
plt.subplot(1, 2, 1)
plt.scatter(data['X1'], data['X2'], s=50, c=data['SVM1 Confidence'], cmap='seismic') #将置信度映射上去，置信度越大说明离边界越远，颜色越深
plt.subplot(1, 2, 2)
plt.scatter(data['X1'], data['X2'], s=50, c=data['SVM2 Confidence'], cmap='seismic') #将置信度映射上去
plt.show()
```

## 1.2 SVM with Gaussian Kernals

### 1.2.1 Gaussian Kernals

```python
def gaussian_kernal(x1, x2, sigma):
    return np.exp(-1 * np.sum(np.power(x1-x2, 2)) / (2 * np.power(sigma, 2)))
```

```python
# 验证高斯kernal的正确性
x1 = np.array([1.0, 2.0, 1.0])
x2 = np.array([0.0, 4.0, -1.0])
sigma = 2

gaussian_kernal(x1, x2, sigma)
```

### 1.2.2 Dataset 2

```python
# 为了生成非线性决策边界，这次更换数据集
raw_data2 = loadmat('data/ex6data2.mat')
print(raw_data2['X'].shape, raw_data2['y'].shape)

data2 = pd.DataFrame(raw_data2['X'], columns=['X1', 'X2'])
data2['Y'] = raw_data2['y']
data2.head()
```

```python
positive_2 = data2[data2['Y'].isin([1])]
negative_2 = data2[data2['Y'].isin([0])]

plt.figure(figsize=(9, 6))
plt.scatter(positive_2['X1'], positive_2['X2'], c='k', marker='+', label='positive')
plt.scatter(negative_2['X1'], negative_2['X2'], c='y', marker='o', label='negative')
plt.legend()
plt.show()
```

```python
svc = svm.SVC(C=100, kernel='rbf', gamma=10, probability=True) # rbf内核，gamma代表rbf内核的核系数,并开启概率估计
svc
```

```python
svc.fit(data2[['X1', 'X2']], data2['Y'])
svc.score(data2[['X1', 'X2']], data2['Y'])
```

the `predict_proba` will give you ndarray (data size, class)  
so if you just want to plot the decision contour of this binary example, choose one class and plot it  
  
Returns the probability of the sample for each class in the model. The columns correspond to the classes in sorted order, as they appear in the attribute classes_.

```python
probability = svc.predict_proba(data2[['X1', 'X2']]) #当然也可以使用前面的置信度
probability[:15, :] #第一列为负类的概率，第二列为正类的概率， 0， 1
```

```python
plt.figure(figsize=(18, 7))
plt.subplot(1, 2, 1)
plt.scatter(data2['X1'], data2['X2'], s=50, c=probability[:, 1], cmap='seismic')
plt.subplot(1, 2, 2)
plt.scatter(data2['X1'], data2['X2'], s=50, c=probability[:, 0], cmap='seismic')
plt.show()
```

## 1.3 Dataset 3

```python
raw_data3 = loadmat('data/ex6data3.mat')
X = raw_data3['X']
y = raw_data3['y'].squeeze()
Xval = raw_data3['Xval']
yval = raw_data3['yval'].squeeze()
print(X.shape, y.shape, Xval.shape, yval.shape)
```

```python
C_values = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30, 100]
sigma_values = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30, 100] #文档中给出

bestscore = -1000000
bestC = 0
bestsigma = 0

for c in C_values:
    for sigma in sigma_values:
        svc = svm.SVC(C=c, kernel='rbf', gamma=sigma)
        svc.fit(X, y)
        score = svc.score(Xval, yval)  #使用 Xval， yval去获得最佳的C 和 sigma
        
        if score > bestscore:
            bestscore = score
            bestC = c
            bestsigma = sigma

print(bestscore, bestC, bestsigma)
```

```python
best_svc = svm.SVC(C=0.3, kernel='rbf', gamma=100, probability=True)
best_svc.fit(X, y)
predict = best_svc.predict(Xval)
predict = predict[predict == yval]
print('Accuracy is ' + str(len(predict) / len(yval) * 100) + '%')

best_confidence = best_svc.decision_function(X)
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=best_confidence, cmap='seismic')
plt.show()
```

# 2. Spam Classification
关于邮件的处理官方已经帮我们完成了 直接跳到2.3小节  

数据直接拿来用就是了，不过邮件的的处理过程大概涉及到NLP方向的吧  
特征提取大意：邮件中的单词如果存在对应的vocabulary_list中则将其下标标上1

```python
spam_train = loadmat('data/spamTrain.mat')
spam_test = loadmat('data/spamTest.mat')
#注：在提取X与y时要知道上面两个数据中的字典的键值是什么名字，需要先打印输出一次，观察它的键值
X_train = spam_train['X']
y_train = spam_train['y'].squeeze()  #将多余维度压缩，后面库函数参数只能传入1维y
X_test = spam_test['Xtest']
y_test = spam_test['ytest'].squeeze()
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

## Train the model
http://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC

```python
svc = svm.SVC() #上面链接中有这个函数的默认设置，C 默认为1，内核默认是rbf 内核系数默认 1/n features
```

```python
svc.fit(X_train, y_train)
predict = svc.predict(X_test)
```

```python
from sklearn import metrics #用于调用评估函数
print(metrics.classification_report(y_test, predict))
```

```python
#当然也可以自己写
predict = svc.predict(X_train)
predict = predict[predict == y_train]
print("Accuracy is " + str(len(predict) / len(y_train) * 100) + "%")
```
